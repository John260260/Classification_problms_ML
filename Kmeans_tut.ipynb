```python
import pandas as pd 
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=[6,4]
plt.style.use('ggplot')
```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/iris.csv')
```

```python
df.head()
```

```python
df2=df.copy()
df2['class'] = df['class'].map({'Iris-setosa':0 ,'Iris-versicolor':1, 'Iris-virginica':2}).copy()
```

```python
### For the purposes of this example, we store feature data from our
### dataframe `df`, in the `X` and `y` arrays. We combine this into
### a feature matrix `X` before entering it into the algorithm.
X = df2[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
```

```python
df2['class'].unique()
```

```python
df2.head(2)
```

```python
Y= df2['class'].values
```

```python

Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()

# When we graph the plot, we see that the graph levels off rapidly after 3 
# clusters, implying that addition of more clusters do not explain much more 
# of the variance in our relevant variable; in this case stock returns.
```

```python
score
```

```python
import pandas
import pylab as pl
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd
```

```python
df.head()
```

```python
Y=df['returns']
X=df['dividendyield']
```

```python
X_norm= (X-X.mean())/(X.max()-X.min())
Y_norm= (Y-Y.mean())/(Y.max()-Y.min())
```

```python
X=X.values
X=X.reshape(-1,1)
Y=Y.values
Y=Y.reshape(-1,1)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()

# When we graph the plot, we see that the graph levels off rapidly after 3 
# clusters, implying that addition of more clusters do not explain much more 
# of the variance in our relevant variable; in this case stock returns.
```

```python
#  pca algorithm is being used to convert data that might be overly dispersed into a set of linear combinations 
# that can more easily be interpreted.
pca=PCA(n_components=1).fit(Y)
pca_d=pca.transform(Y)
pca_c=pca.transform(X)
```

```python
kmeans=KMeans(n_clusters=3)
kmeansoutput=kmeans.fit(Y)
kmeansoutput
```

```python
np.unique(kmeansoutput.labels_)
```

```python
plt.figure('3 cluster K-Means')
plt.scatter(pca_c[:,0],pca_d[:,0], c=kmeansoutput.labels_)
plt.xlabel('Dividend Yield')
plt.ylabel('Returns')
plt.title('3 Cluster K-Means')
plt.show()
```

#### numpy.linalg.norm(x, ord=None, axis=None, keepdims=False)
Matrix or vector norm.

This function is able to return one of 8 different matrix norms, or one of an infinite number of vector norms, depending on the value of the __ord__ parameter.

__axis__

- If axis is an integer, it specifies the axis of x along which to compute the vector norms. 

- If axis is a 2-tuple, it specifies the axes that hold 2-D matrices, and the matrix norms of these matrices are computed. 

- If axis is None then either a vector norm (when x is 1-D) or a matrix norm (when x is 2-D) is returned.
